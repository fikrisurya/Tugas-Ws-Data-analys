```python
ppdb = pd.read_excel('Data PPDB (2).xlsx')
ppdb.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='Jenkel', data=ppdb)
plt.title("Grafik jenkel")
plt.xlabel("Jenkel")
plt.ylabel("Jumlah")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='Transport', data=ppdb)
plt.title("Grafik transport")
plt.xlabel("transportasi")
plt.ylabel("Jumlah")
plt.show()
```

```python
data_diatas_6_tahun = ppdb[ppdb["Umur"]> 6 ]

# Menghitung jumlah peserta didik berdasarkan jenis kelamin
count_by_gender = data_diatas_6_tahun['Jenkel'].value_counts()

# Visualisasi dalam bentuk bar chart
count_by_gender.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Jumlah Peserta Didik di Atas 6 Tahun Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Peserta Didik')
plt.show()
```

```python
# Memfilter data untuk peserta didik perempuan dengan usia di atas 6 tahun
ppdb_usia_jenkel = ppdb[(ppdb['Umur'] > 6) & (ppdb['Jenkel'] == 'P')]

# Menggunakan seaborn untuk membuat count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Umur', data=ppdb_usia_jenkel)
plt.title('Distribusi Usia Peserta Didik Perempuan di Atas 6 Tahun')
plt.xlabel('Usia')
plt.ylabel('Jumlah Peserta Didik')
plt.show()
```
